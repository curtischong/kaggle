```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import Imputer

# Read Data
data = pd.read_csv("train.csv")
#test = pd.read_csv("test.csv")
```

```python
data.iloc[2].Ticket #I'll worry about tickets leter. The forums say it's good for locating families
```

```python
def returnStr(x):
    saved = ""
    for idx in x:
        if idx.isdigit():
            return saved
        saved = saved + idx
    return saved

data.drop(['Embarked'],inplace=True,errors="ignore",axis=1) # I don't think the port that they entered at is important
if "Cabin" in data:
    data.Cabin = data.Cabin.fillna("0")
    data["Deck"] = data.Cabin.map(lambda x: x[0])
    data["Room_num"] = data.Cabin.map(lambda x: x[1:]) # do not one hot encode this
data.drop(['Cabin'],inplace=True,errors="ignore",axis=1)
data['Age'] = data['Age'].fillna(data['Age'].median())
data.Ticket = data.Ticket.map(lambda x: returnStr(x))
data
```

```python
#credit (Ahmed BESBES): https://ahmedbesbes.com/how-to-score-08134-in-titanic-kaggle-challenge.html
titles = set()
for name in data['Name']:
    titles.add(name.split(',')[1].split('.')[0].strip())

Title_Dictionary = {
    "Capt": "Officer",
    "Col": "Officer",
    "Major": "Officer",
    "Jonkheer": "Royalty",
    "Don": "Royalty",
    "Sir" : "Royalty",
    "Dr": "Officer",
    "Rev": "Officer",
    "the Countess":"Royalty",
    "Mme": "Mrs",
    "Mlle": "Miss",
    "Ms": "Mrs",
    "Mr" : "Mr",
    "Mrs" : "Mrs",
    "Miss" : "Miss",
    "Master" : "Master",
    "Lady" : "Royalty"
}

def get_titles():
    # we extract the title from each name
    combined['Title'] = combined['Name'].map(lambda name:name.split(',')[1].split('.')[0].strip())
    
    # a map of more aggregated title
    # we map each title
    combined['Title'] = combined.Title.map(Title_Dictionary)
    return combined
```

```python
from sklearn.preprocessing import LabelEncoder
# I have to replace the original lables and do one hot encoding
data[['Sex',]].apply(LabelEncoder().fit_transform)
```

```python
#cols_to_use = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch','Fare','Cabin','Embarked']

cols_to_use = ['Fare','Sex']
X = data[cols_to_use]
y = data.Survived
train_X, test_X, train_y, test_y = train_test_split(X, y)
```

```python
my_pipeline = make_pipeline(Imputer(), RandomForestRegressor())
my_pipeline.fit(train_X, train_y)
predictions = my_pipeline.predict(test_X)
```

```python
# Read the test data
test = pd.read_csv('test.csv')
# Treat the test data in the same way as training data. In this case, pull same columns.
test_X = test[cols_to_use]
# Use the model to make predictions
predicted = my_pipeline.predict(test_X)
# We will look at the predicted prices to ensure we have something sensible.
print(predicted)
newArr = []
for x in predicted:
    newArr.append(int(round(x)))
```
