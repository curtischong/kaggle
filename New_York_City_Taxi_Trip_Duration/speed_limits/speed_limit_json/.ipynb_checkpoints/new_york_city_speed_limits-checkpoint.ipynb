```python
#http://www.nyc.gov/html/dot/html/about/vz_datafeeds.shtml
import pandas as pd
import numpy as np
import json
import geocoder
from geopy.geocoders import Nominatim
geolocator = Nominatim()
```

```python
speed_limit_bronx = (pd.read_json("speed_limit_bronx.json")["features"])
speed_limit_brooklyn = (pd.read_json("speed_limit_brooklyn.json")["features"])
speed_limit_manhattan = (pd.read_json("speed_limit_manhattan.json")["features"])
speed_limit_queens = (pd.read_json("speed_limit_queens.json")["features"])
speed_limit_statenisland = (pd.read_json("speed_limit_statenisland.json")["features"])
```

```python
type(speed_limit_bronx)
```

```python
all_speeds = []
for item in speed_limit_bronx:
    all_speeds.append((item,"Bronx"))
for item in speed_limit_brooklyn:
    all_speeds.append((item,"Brooklyn"))
for item in speed_limit_manhattan:
    all_speeds.append((item,"Manhattan"))
for item in speed_limit_queens:
    all_speeds.append((item,"Queens"))
for item in speed_limit_statenisland:
    all_speeds.append((item,"Staten Island"))
```

```python
speeds = []
for limit in all_speeds:
    coords = limit[0]["geometry"]["coordinates"]
    properties = limit[0]["properties"]
    speeds.append([coords[0][1],coords[0][0],coords[1][1],coords[1][0],properties["street"].lower(), properties["postvz_sl"], properties["postvz_sg"].lower(),limit[1]])
```

```python
df = pd.DataFrame(speeds,columns=["lat1","lon1","lat2","lon2","street","speed","signed","region"])
```
